```python
import nltk
from googletrans import Translator
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
#YOUTUBE COMMENT EXTRACTOR
lista_comments=[]
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver import ActionChains
import time
import datetime
from bs4 import BeautifulSoup
import random
def sl(x1):
    time.sleep(x1)

driver = webdriver.Chrome("C:\\Users\\asus\\Downloads\\chromedriver_win32\\chromedriver.exe")

driver.get("https://www.youtube.com/watch?v=F6KwvxCxqqA&ab_channel=BBCNews")
sl(12)
try:
    cookie_button=WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,'//yt-formatted-string[contains(text(),"Skip trial")]')))
    time.sleep(1.3)
    cookie_button.click()
    time.sleep(2)
except:
    pass
driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
sl(random.randint(8,12))
for i in range(5):
    driver.find_element_by_tag_name('body').send_keys(Keys.END)
    sl(random.randint(8,13))

pagina = driver.page_source
#aqui sai uma string
pagina_bs = BeautifulSoup(pagina,'html.parser')

textos=pagina_bs.find_all('yt-formatted-string',attrs={'id': 'content-text'})
for i in textos:
    print (i.string)
    lista_comments.append(i.string)
    print ('\n\n')
# while True:
#     last_height = driver.execute_script("return document.body.scrollHeight")

#     new_height = driver.execute_script("return document.body.scrollHeight")
#     if last_height==new_height:
#         break

lista_comments1=[str(i) for i in lista_comments if len(str(i))>3]
df=pd.DataFrame(lista_comments1,columns=['Comments'])
def sentimento(comment):
    if comment>0:
        return 'Positive'
    elif comment<0:
        return 'Negative'
    else:
        return 'Neutral'
def resultado(df1):
    print ('NEUTRAL COMMENTS: '+str(len(df1[df1['Sentiment']=='Neutral'])))
    print ('POSITIVE COMMENTS: '+str(len(df1[df1['Sentiment']=='Positive'])))
    print ('NEGATIVE COMMENTS: '+str(len(df1[df1['Sentiment']=='Negative'])))

df['Compound']=df['Comments'].apply(lambda x:sid.polarity_scores(x)['compound'])
df['Sentiment']=df['Compound'].apply(lambda x:sentimento(x))
resultado(df)

drive.close()
```
